## STEP 1: Load and Filter data

For this analysis, we want to look only at a relevent range of years and only years that are in teh past.  So, let's first take a look at the data between 1986 and 2014 (inclusive of those dates).

At the end of this step, you should have a data frame loaded from the condemnations file and filtered down to only the data between 1986 and 2014.  Put that into a dataframe named `condemnations`

```python
%matplotlib inline
import pandas as pd

condemnations = pd.read_csv('/content/condemn.txt')
condemnations.shape
```

```python
condemnations = pd.read_csv('/content/condemn.txt', low_memory=False)
condemnations.shape
condemnations['InspectDate'] = pd.to_datetime(condemnations['InspectDate'])
condemnations = condemnations[(condemnations['InspectDate'] >= '1986-01-01') &
                              (condemnations['InspectDate'] <= '2014-12-31')]
```

## STEP 2: Pivot and Summarize

In this step we want to pivot our data so that we have one row per year, a column for each value that appears in the `Status` column, and a count of how many condemnations occurred of each type in each year.  Put your summary pivot table into a variable called `by_year`.

**Some helpful notes:**
* For some reason, Pandas is just kind of stubborn about count being the aggregate function.  You may need to tell it to `np.ma.count` specifically.  I've imported Numpy as np for you.

```python
import numpy as np
import matplotlib.pyplot as plt
# Extract the year from the 'InspectDate' column
condemnations['Year'] = condemnations['InspectDate'].dt.year
```

```python
# Pivot the data with counts for each Status type by year
by_year = condemnations.pivot_table(index='Year',
                                    columns='Status',
                                    values='InspectDate',
                                    aggfunc=np.ma.count,
                                    fill_value=0)
```

```python
assert(by_year.shape == (29, 4))
assert(list(by_year.columns) == ['B', 'L', 'R', 'W'])
assert(list(by_year.sum()) == [10572, 4966, 98, 9670])
```

```python
import matplotlib.pyplot as plt

# Plotting a stacked area plot
plt.figure(figsize=(12, 6))
plt.fill_between(by_year.index, by_year.sum(axis=1), color='lightgrey', label='Total Condemnations')

for column in by_year.columns:
    plt.fill_between(by_year.index, by_year[column].cumsum(),
                     label=column, alpha=0.6)

# Add labels and legend
plt.title("Condemnations by Year and Status")
plt.xlabel("Year")
plt.ylabel("Count of Condemnations")
plt.legend(title="Status", loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Show the plot
plt.show()
```

This area chart displays the total count of condemnations over time, divided by categories labeled "B," "L," "R," and "W." The x-axis shows the years from 1985 to 2015, while the y-axis represents the count of condemnations. Each colored area in the chart corresponds to a different status category, and the areas stack on top of each other to show the cumulative total condemnations over the years.

From the chart, we can observe a general upward trend in condemnations, with significant increases from the early 2000s to 2015. The category "W" (in pink) represents the largest portion of condemnations, followed by "R" (in red) and other smaller contributions from "L" and "B." The total number of condemnations reaches over 10,000 by 2015.

```python
import pandas as pd

#  Load the data
meetings = pd.read_csv('/content/meetings.csv')

#  Extract 'startDate' from 'startTime' (first 10 characters for date part)
meetings['startDate'] = meetings['startTime'].str[0:10]

meetings['Period'] = None
meetings.loc[(meetings['startDate'] >= '2020-02-17') & (meetings['startDate'] <= '2020-02-28'), 'Period'] = 'Before'
meetings.loc[(meetings['startDate'] >= '2020-03-16') & (meetings['startDate'] <= '2020-03-27'), 'Period'] = 'During'

# Filter to include only rows in either "BEFORE" or "DURING" period
meetings = meetings.dropna(subset=['Period'])
```

```python
summary = meetings.pivot_table(index='userName', columns='Period', values='meeting_uuid', aggfunc='count', fill_value=0)

#  Rename columns
summary.columns = ['Before', 'During']

# Calculate percent change (pctChange) from "Before" to "During"
summary['pctChange'] = ((summary['During'] - summary['Before']) / summary['Before']).replace([float('inf'), -float('inf')], 0).fillna(0)
```
