### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
import pandas as pd

# Read the data file
file_path = '/content/Medicaid_Coverage_Of_Cessation_Treatments_And_Barriers_To_Treatments (2).csv'
data = pd.read_csv(file_path)

# Calculate the number of unique values in each column
unique_counts = data.nunique()

# Find the column with the greatest number of distinct values
most_diverse_column = unique_counts.idxmax()
most_diverse_count = unique_counts.max()

# Identify the column with the next most unique values
sorted_unique_counts = unique_counts.sort_values(ascending=False)
next_most_diverse_column = sorted_unique_counts.index[1]
next_most_diverse_count = sorted_unique_counts.iloc[1]

# Print the results
print("Unique values per column:\n", unique_counts)
print("\nMost diverse column:", most_diverse_column, "with", most_diverse_count, "unique values")
print("Next most diverse column:", next_most_diverse_column, "with", next_most_diverse_count, "unique values")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt

# Plot a histogram for the 'Summary_AltValue' column
plt.figure(figsize=(10, 6))
plt.hist(data['Summary_AltValue'], bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Summary_AltValue', fontsize=14)
plt.xlabel('Summary_AltValue', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/Medicaid_Coverage_Of_Cessation_Treatments_And_Barriers_To_Treatments (2).csv'
data = pd.read_csv(file_path)

# Filter for specific barriers of interest
selected_barriers = ["Co-Payments Required", "Prior Authorization Required"]
filtered_data = data[data['Sub_Measure'].isin(selected_barriers)]

# Group by Year and Sub_Measure to count occurrences
barriers_trend = filtered_data.groupby(['Year', 'Sub_Measure']).size().unstack()

# Plot trends for the selected barriers
plt.figure(figsize=(12, 6))
barriers_trend.plot(ax=plt.gca(), marker='o')
plt.title('Trends in Selected Barriers to Medicaid Cessation Treatments (2008-2024)')
plt.xlabel('Year')
plt.ylabel('Frequency of Barrier')
plt.legend(title='Barrier Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```
